#**Singly Linked List :**
**Python code for Singly Linked List :**

```
class Node: # Class for creating a node
    def __init__(self, data):
        self.next = None
        self.data = data

class SLL: # Class for creating a Singly Linked List
    def __init__(self):
        self.head = None

    def insertBeg(self, data): # Method for insertion at the beginning of a list
        newnode = Node(data)

        if self.head == None:
            self.head = newnode
            newnode.next = None
        else:
            newnode.next = self.head
            self.head = newnode

    def insertEnd(self, data):# Method for insertion at the end of a list
        newnode = Node(data)

        if self.head is None:
            self.head = newnode

        else:
            ptr = self.head
            while ptr.next is not None:
                ptr = ptr.next
            ptr.next = newnode

    def insertBeforeNode(self, data1, data2):# Method for insertion before a given node in a list
        newnode = Node(data2)

        if self.head is None:
            self.head = newnode

        else:
            ptr = self.head
            while ptr.next.data is not data1:
                ptr = ptr.next
            newnode.next = ptr.next
            ptr.next = newnode

    def insertAfterNode(self, data1, data2):# Method for insertion after a given node in a list
        newnode = Node(data2)

        if self.head is None:
            self.head = newnode

        else:
            ptr = self.head
            while ptr.data is not data1:
                ptr = ptr.next
            newnode.next = ptr.next
            ptr.next = newnode

    def deleteBeg(self):# Method for deleting a node from the beginning of the list
        if self.head is None:
            print("Underflow!")
        else:
            self.head = self.head.next

    def deleteEnd(self):# Method for deleting a node from the end of the list
        if self.head is None:
            print("Underflow!")
        else:
            ptr = self.head
            while ptr.next.next is not None:
                ptr = ptr.next
            ptr.next = None

    def deleteMid(self, data):# Method for deleting a particular node of the list
        if self.head is None:
            print("Underflow!")
        else:
            ptr = self.head
            while ptr.next.data != data:
                ptr = ptr.next
            ptr.next = ptr.next.next

    def printList(self):# Method for printing the given list
        ptr = self.head
        while ptr is not None:
            print(ptr.data, end = " ")
            ptr = ptr.next
        print("\n")
sll = SLL()
sll.head = Node(5)
sll.insertEnd(6)
sll.insertEnd(8)
sll.insertEnd(9)
sll.insertEnd(11)
sll.printList()
sll.insertBeforeNode(8,7)
sll.insertAfterNode(9,10)
sll.printList()
sll.deleteBeg()
sll.deleteEnd()
sll.deleteMid(9)
sll.printList()
```

#**Stack as Singly Linked List :**
**Python code for Stack implemented as a Singly Linked List :**

```
class Node: # Class for creating a Node
    def __init__(self, data):
        self.next = None
        self.data = data

class Stack: # Class for creating a Stack
    def __init__(self):
        self.top = None

    def push(self, data): # Method for pushing a node into the stack
        newnode = Node(data)

        if self.top is None:
            self.top = newnode
        else:
            newnode.next = self.top
            self.top = newnode

    def pop(self): # Method for popping a node from the stack
        if self.top is None:
            print("Underflow!")
        else:
            self.top = self.top.next

    def printStack(self): # Method for printing the stack
        if self.top is None:
            print("Underflow!")
        else:
            print(self.top.data, "<--- top")
            ptr = self.top.next
            while ptr is not None:
                print(ptr.data)
                ptr = ptr.next
            print("\n")

stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)
stack.printStack()
stack.pop()
stack.printStack()
stack.pop()
stack.printStack()
```

#**Queue as a Singly Linked List :**
**Python code for Queue implemented as a Singly Linked List :**

```
class Node: # Class for creating a Node
    def __init__(self, data):
        self.next = None
        self.data = data

class Queue: # Class for creating a Queue
    def __init__(self):
        self.first = None
        self.last = None

    def insert(self, data): # Method for inserting element into the Queue
        newnode = Node(data)

        if self.last is None:
            self.first = self.last = newnode

        else:
            self.last.next = newnode
            self.last = newnode

    def delete(self): # Method for deleting element from Queue
        if self.first is None:
            print("Underflow!")

        else:
            self.first = self.first.next

    def printQueue(self): # Method for printing the Queue
        if self.first is None and self.last is None:
            print("Underflow!")

        else:
            ptr = self.first
            while ptr is not None:
                print(ptr.data, end = " ")
                ptr = ptr.next
            print("\n")

q = Queue()
q.insert(5)
q.printQueue()
q.insert(6)
q.printQueue()
q.insert(7)
q.insert(8)
q.insert(9)
q.printQueue()
q.delete()
q.printQueue()
```

#**Deque as a Singly Linked List :**
**Python code for Deque implemented as a Singly Linked List :**

```
class Node: # Class for creating a Node
    def __init__(self, data):
        self.data = data
        self.next = None

class Deque: # Class for creating a Deque
    def __init__(self):
        self.first = None
        self.last = None

    def insertFront(self, data): # Method for inserting node in front of the deque
        newnode = Node(data)

        if self.first is None:
            self.first = self.last = newnode

        else:
            newnode.next = self.first
            self.first = newnode

    def insertEnd(self, data): # Method for inserting node at the end of deque
        newnode = Node(data)

        if self.last is None:
            self.first = self.last = newnode

        else:
            self.last.next = newnode
            self.last = newnode

    def deleteFront(self): # Method for deleting node from the front of the deque
        if self.first is None:
            print("Underflow!")
        else:
            self.first = self.first.next

    def deleteEnd(self): # Method for deleting node from the end of deque
        if self.last is None:
            print("Underflow!")
        else:
            ptr = self.first
            while ptr.next.next is not None:
                ptr = ptr.next
            ptr.next = None

    def printDeque(self): # Method for printing the deque
        if self.first is None:
            print("Underflow!")

        else:
            ptr = self.first
            while ptr is not None:
                print(ptr.data, end = " ")
                ptr = ptr.next
            print("\n")

dq = Deque()
dq.insertFront(5)
dq.insertFront(6)
dq.printDeque()
dq.insertFront(7)
dq.insertEnd(8)
dq.printDeque()
dq.deleteFront()
dq.printDeque()
dq.deleteEnd()
dq.printDeque()
```

#**Queue using Stack :**
**Python code for Queue implemented using Stack :**

```
class Queue: # Class for creating a Queue using Stacks
    def __init__(self):
        self.inbox = Stack()
        self.outbox = Stack()

    def is_empty(self): # Method for checking if queue is empty or not
        return (self.inbox.is_empty() and self.outbox.is_empty())

    def enqueue(self, data): # Method to insert node in the queue
        self.inbox.push(data)

    def dequeue(self): # Method for deleting node from queue
        if self.outbox.is_empty():
            while not self.inbox.is_empty():
                popped = self.inbox.pop()
                self.outbox.push(popped)
            self.outbox.pop()
        while not self.outbox.is_empty():
            popped = self.outbox.pop()
            self.inbox.push(popped)

    def printQueue(self): # Method for printing Queue
        self.inbox.printStack()

class Stack: #Class for creating Stack, to be used subsequently for creating a Queue
    # Functions perform same tasks as above respectively
    def __init__(self):
        self.items = []

    def is_empty(self):
        return self.items == []

    def push(self, data):
        self.items.append(data)

    def pop(self):
        return self.items.pop()

    def printStack(self):
        print(self.items)

queue = Queue()

queue.enqueue(5)
queue.enqueue(6)
queue.printQueue()
queue.enqueue(7)
queue.printQueue()
queue.dequeue()
queue.printQueue()
```

#**Stack using Queue :**
**Python code for Stack implemented using Queue :**

```
class Stack: # Class for creating Stack using Queue
    def __init__(self):
        self.q = Queue()

    def is_empty(self): # Method to check whether the stack is empty or not
        return self.q.is_empty()

    def push(self, data): # Method to push a node into the stack
        self.q.enqueue(data)

    def pop(self): # Method to pop a node from the stack
        for _ in range(self.q.get_size() - 1):
            dequeued = self.q.dequeue()
            self.q.enqueue(dequeued)
        return self.q.dequeue()

    def printStack(self): # Method to print the stack
        self.q.printQueue()


class Queue: # Class for creating a Queue, to be subsequently used to create Stack
    # Functions perform same tasks as above respectively
    def __init__(self):
        self.items = []
        self.size = 0

    def is_empty(self):
        return self.items == []

    def enqueue(self, data):
        self.size += 1
        self.items.append(data)

    def dequeue(self):
        self.size -= 1
        return self.items.pop(0)

    def get_size(self):
        return self.size

    def printQueue(self):
        print(self.items)


stack = Stack()

stack.push(5)
stack.push(6)
stack.printStack()
stack.push(7)
stack.printStack()
stack.pop()
stack.printStack()
```
